## Queations

### 1. Convert the height and weight columns to numerical forms

```python
print(df1.dtypes)
```

```python
height_data = df1['Height']
print(height_data)
```

```python
def convert_to_cm(height):
    try:
        feet, inches = height.split("'")
        return int(feet) * 30.48 + int(inches[:-1]) * 2.54
    except: 
        return None
    
df1['Height'] = df1['Height'].apply(convert_to_cm)
print(df1['Height'])
```

```python
weight_data = df1['Weight']
print(weight_data)
```

```python
def convert_to_kg(weight):
    try: 
        return int(weight.replace('lbs',''))
    except:
        return None
    
df1['Weight'] = df1['Weight'].apply(convert_to_kg)
print(df1['Weight'])
```

### 2. Remove the unnecessary newline characters from all columns that have them.

#### 'Hits' column has new line characters which should be removed.

### 3. Based on the 'Joined' column, check which players have been playing at a club for more than 10 years!

```python
# Extracting year, month and day from the 'Joined' column
df1['Join_Year'] = pd.to_datetime(df1['Joined'], format="%b %d, %Y").dt.year
df1['Join_Month'] = pd.to_datetime(df1['Joined'], format="%b %d, %Y").dt.month
df1['Join_Day'] = pd.to_datetime(df1['Joined'], format="%b %d, %Y").dt.day
```

### 4. 'Value', 'Wage' and "Release Clause' are string columns. Convert them to numbers. For eg, "M" in value column is Million, so multiply the row values by 1,000,000, etc.

```python
df1.Value
```

```python
def convertMoney(money: str):
    value = pd.to_numeric(re.findall(pattern=r"\d+\.?\d+|\d", string=money))[0]
    if value == np.nan: value = 0
    if "M" in money:
        value *= 1000000
    elif "K" in money:
        value *= 1000
    return str(int(value))
    

df1.Value = pd.to_numeric(df1.Value.apply(convertMoney))
df1.Wage = pd.to_numeric(df1.Wage.apply(convertMoney))
df1['Release Clause'] = pd.to_numeric(df1['Release Clause'].apply(convertMoney))
```

### 5. Some columns have 'star' characters. Strip those columns of these stars and make the columns numerical

```python
df1[['W/F','SM','IR']]
```

```python
df1['W/F'] = df1['W/F'].apply(lambda x: re.sub(pattern="[^\d]", repl="",string=x))
df1['SM'] = df1['SM'].apply(lambda x: re.sub(pattern="[^\d]", repl="",string=x))
df1['IR'] = df1['IR'].apply(lambda x: re.sub(pattern="[^\d]", repl="",string=x))
```
